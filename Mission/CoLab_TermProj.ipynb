## Transcript count table 마련하기

제공된 alignment와 annotation 파일들을 이용해서 transcript별 read count를 구해봅니다. 실제 연구에서는 multi-mapping 등을 고려해야 하지만, 여기서는 단순화해서 모두 무시합니다. 데이터가 있는 곳으로 이동해서 작업합시다.

````python
%cd /data/project/jihye/2021-1/BioinfoPractice1/TermProject/binfo1-datapack1
````

````python
!featureCounts -a gencode.gtf -o read-counts.txt *.bam 
````

````python
!ls
````

결과를 pandas로 읽어와 봅니다.

````python
import pandas as pd
cnts = pd.read_csv('read-counts.txt', sep='\t', comment='#')
cnts.head()
````

````python
cnts.columns
````

이제 한 번 대충 그림을 그려 봅니다.

````python
cnts['clip_enrichment'] = cnts['CLIP-35L33G.bam'] / cnts['RNA-control.bam']
cnts['rden_change'] = (cnts['RPF-siLin28a.bam'] / cnts['RNA-siLin28a.bam']) / (cnts['RPF-siLuc.bam'] / cnts['RNA-siLuc.bam'])
cnts.head()
````

````python
def rm_after_dot(str):
    ans, temp = str.split('.')
    return ans
````

````python
cnts['Geneid'] = cnts['Geneid'].apply(rm_after_dot)
cnts
````

````python
cnts.rename(columns = {"Geneid": "gene_id"}, inplace=True)
cnts
````

````python
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots(1, 1, figsize=(5, 5))
ax.scatter(np.log2(cnts['clip_enrichment']),
           np.log2(cnts['rden_change']))
````

이 그림에는 문제가 많이 있습니다. 논문의 그림처럼 한 번 만들어 봅시다~

## Protein localization 반영하기

논문 그림에서는 protein localization을 반영해서 색으로 나타냈습니다. 이것을 한 번 구현해 봅시다. 각 gene product의 localization을 따로 구하려면 좀 귀찮기 때문에 미리 모아 두었습니다. UniProt에서 매핑된 데이터를 토대로 아주 명확한 것만 일부 정리한 것입니다. 이것도 마찬가지로 처음엔 에러가 나기도 하니까, 다시 실행해보면 거의 잘 됩니다~

````python
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
mouselocal = pd.read_csv('https://hyeshik.qbio.io/binfo/mouselocalization-20210507.txt', sep='\t')
mouselocal.head()
````

이제 이 localization 데이터와 위에서 만든 scatter를 결합해서 논문 그림과 비슷하게 만들어 봅시다.

````python
cnts.set_index('gene_id', inplace=True)
mouselocal.set_index('gene_id', inplace=True)
````

````python
merged = mouselocal.join(cnts, how='inner')
merged.head()
````

````python
merged['clip_enrichment'] = merged['CLIP-35L33G.bam'] / merged['RNA-control.bam']
merged['rden_change'] = (merged['RPF-siLin28a.bam'] / merged['RNA-siLin28a.bam']) / (merged['RPF-siLuc.bam'] / merged['RNA-siLuc.bam'])
merged.head()
````

````python
merged.reset_index(inplace=True)
merged
````

````python
merged.reset_index(drop=True, inplace=True)
merged
````

````python
merged['type'][:,]
````

````python
plt.plot(figsize=(5,5))
colors = {'nucleus':'blue','integral membrane':'red','cytoplasm':'green'}
plt.scatter(np.log2(merged['clip_enrichment']), # x
            np.log2(merged['rden_change']), # y
            s=2, # marker size
            c=merged['type'].map(colors),# marker color
            label=merged['type']
           ) 

plt.xlabel(r'LIN28A CLIP enrichment ($log_2$)')
plt.ylabel(r'Ribosome density change upon LIN28a knockdown ($log_2$)')
plt.legend(['Nucleus','Integral Membrane','Cytoplasm'])
plt.axis([-6, 4, -2, 3])     # X, Y축의 범위: [xmin, xmax, ymin, ymax]
````

````python
groups = merged.groupby('type')

 

fig, ax = plt.subplots(1, 1, figsize=(5, 5))

for name, group in groups:

    ax.plot(group.petal_length, 

            group.petal_width, 

            marker='o', 

            linestyle='',

            label=name)

ax.legend(fontsize=12, loc='upper left') # legend position

plt.title('Scatter Plot of iris by matplotlib', fontsize=20)

plt.xlabel('Petal Length', fontsize=14)

plt.ylabel('Petal Width', fontsize=14)

plt.show()



출처: https://rfriend.tistory.com/414?category=675918 [R, Python 분석과 프로그래밍의 친구 (by R Friend)]
````
